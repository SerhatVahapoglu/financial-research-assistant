```python
import os
from dotenv import load_dotenv
from langchain_openai import ChatOpenAI
from qdrant_client import QdrantClient
import redis

# .env dosyasındaki değişkenleri sisteme yükler
load_dotenv()
print("✅ Çevre değişkenleri yüklendi!")
```

```python
# LLM ile ufak bir sohbet başlatarak API anahtarımızın çalışıp çalışmadığını test ediyoruz
try:
    llm = ChatOpenAI(model="gpt-3.5-turbo", temperature=0)
    response = llm.invoke("Merhaba, sadece 'Selam' diyerek cevap ver.")
    print("✅ OpenAI Bağlantısı Başarılı! Gelen Cevap:", response.content)
except Exception as e:
    print("❌ OpenAI Hatası:", e)
```

```python
# Vektör veritabanımızın ayakta olup olmadığını kontrol ediyoruz
try:
    qdrant_url = os.getenv("QDRANT_URL", "http://localhost:6333")
    client = QdrantClient(url=qdrant_url)
    
    # Sadece sunucu durumunu istiyoruz, koleksiyon oluşturmaya gerek yok
    collections = client.get_collections()
    print("✅ Qdrant Bağlantısı Başarılı! Hazır durumda.")
except Exception as e:
    print("❌ Qdrant Hatası (Docker çalışıyor mu?):", e)
```

```python
# İleride MCP ve Web Researcher kısmında yedeğimiz olacak Redis'i test ediyoruz
try:
    redis_url = os.getenv("REDIS_URL", "redis://localhost:6379")
    r = redis.from_url(redis_url)
    
    # Basit bir ping atıyoruz
    if r.ping():
        print("✅ Redis Bağlantısı Başarılı!")
except Exception as e:
    print("❌ Redis Hatası (Docker çalışıyor mu?):", e)
```
